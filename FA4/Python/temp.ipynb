```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function
def func(U):
    return (math.factorial(U**2))**U

# Generate values for U
U_values = np.arange(1, 5)
Y_values = [func(U) for U in U_values]

# Plot the function
plt.plot(U_values, Y_values, marker='o')
plt.xlabel('U')
plt.ylabel('(U^2 !)^U')
plt.title('Plot of (U^2 !)^U with increasing U from 1 to 10')
plt.grid(True)
plt.show()
```

```python
# Calculate new Y_values for U * (2^(U^2) - 1)
new_Y_values = U_values * (2**(U_values**2) - 1)

# Plot the new function
plt.plot(U_values, new_Y_values, marker='o')
plt.xlabel('U')
plt.ylabel('U * (2^(U^2) - 1)')
plt.title('Plot of U * (2^(U^2) - 1) with increasing U from 1 to 4')
plt.grid(True)
plt.show()
```
